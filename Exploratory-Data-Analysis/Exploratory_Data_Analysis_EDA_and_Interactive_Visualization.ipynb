```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

years=range(2000,2012)
yield_apples=apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
yield_oranges= [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
sns.set_style('darkgrid')
plt.plot(years,yield_apples,marker='o',c='r')
plt.plot(years,yield_oranges,marker='x',c='b')
plt.title('Crop yields in Kanto')
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')
plt.legend(['Apples','Oranges'])

```

```python
# we can use the plt.hist() to create a histogram

# load data into a Pandas datafram
flowers_df=sns.load_dataset("iris")
flowers_df.sepal_width
```

```python
plt.title("Distributions of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

```python
# we can control the numbers of bins of the size of each one using the bins(intervals) argument
# specifying the number of bins
plt.hist(flowers_df.sepal_width,bins=5);
```

```python
import numpy as np
# specifying the boundaries of each bin
plt.hist(flowers_df.sepal_width,bins=np.arange(2,5,0.25));
```

```python
# bins of unequal sizes
plt.hist(flowers_df.sepal_width,bins=[1,3,4,4.5]);
```

```python
# multiple histograms we can draw multiple histograms in a single chart.
# we can reduce each histogram's opacity so that one histogram'bars don't hide the others

setosa_df=flowers_df[flowers_df.species=='setosa']
versicolor_df=flowers_df[flowers_df.species=='versicolor']
virginica_df=flowers_df[flowers_df.species=='virginica']

plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));

```

```python
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],
         bins=np.arange(2,5,0.25),
         stacked=True);
plt.legend(['Setosa','Versicolor','Virginica']);
```

```python
# Heatmap - A heatmap is used to visualize 2 dimensional data like a matrix or a table using colors.
# we"ll use another sample dataset from Seaborn called ,flights, to visualize monthly passenger football at an airport over 12 years.
flights_df=sns.load_dataset("flights").pivot("month","year","passengers")
flights_df
# flights_df is matrix with one row for each month and one column for each year.
# The values show the number of passengers(in thousands) that visited the airport in a specific
# month of a yars. We can use the sns.heatmap function to visualize the football at the airport.
```

```python
plt.title("No. of Passengers(1000s)")
# use heatmap() function for the
sns.heatmap(flights_df);

'''The brighter colors indicate a higher footfall at the airport.
    By looking at the graph,we can infer two things:
     * the footfall at the airport in any given year tends to be the highest around july and august.
     * the footfall at the airport in any given month tends to grow year by year'''
```

```python
''' We can also display the actual values in each block by specifying annot=True and using
    the cmap argument to change the color palette'''

plt.title("No of Passengers(1000s)")
sns.heatmap(flights_df,fmt="d",annot=True,cmap='Blues');
```

```python
# Use of Plotly library
'''Unlike  Matplotlib and Seaborn which create static images,
   plotly renders an html document and uses javascript under the hood
   to enable interactivity. Plotly also offers a large selection of chart
   types of choose from
'''



import pandas as pd
import numpy as np
import plotly.express as px

population_csv_url='https://gist.githubusercontent.com/aakashns/bbd36fbd7c0be266f0c875ad2006a9fd/raw/1763ee47c8919995c4115bb063c99511ced34712/population.csv'
population_df=pd.read_csv(population_csv_url,index_col='Year')  # here we are setting  the index_col= year
population_df[0:5]

```

```python
''' let's use px.line to create a line chart showing
    the population of hungary from 1960 to 2019.
'''
px.line(population_df['Hungary'],title="Population")
```

```python
fig=px.line(population_df['Hungary'])
```

```python
# set Axis and legend labels
# for updating the layout
fig.update_layout(
    title='Year-Wise Population',
    xaxis_title='Year',
    yaxis_title='Population',
    legend_title='Country',
    plot_bgcolor='orange',
    font=dict(
        family="Arial",
        size=14,
        color='blue'
    )
)
# Start the y axis from 0
fig.update_yaxes(rangemode='tozero')
```

```python
europe_df=population_df[['Hungary','Czech Republic','Switzerland']]
europe_df.head()
```

```python
fig=px.line(europe_df,title='Population',
            color_discrete_sequence=['aquamarine','cornflowerblue','goldenrod'])
fig.update_layout(yaxis_title='Population',
                  legend_title='Countries',
                  font_size=14)

fig.update_yaxes(rangemode='tozero')




```

```python
'''switching from a line chart to a bar chart
   is simply a matter replacing plt.line with px.bar
 '''

px.bar(population_df[['Bangladesh','Pakistan']],
        title="Population",
        barmode='group')
```

```python
px.bar(population_df[['India','China']],
       title='Population',
        )
```

```python
fig=px.bar(population_df[['India','China']],
                 title='Population',
                 barmode='group' )

fig.update_layout(yaxis_title='Population',
                  legend_title='Countries',
                  font_size=14,
                  legend_itemclick='toggle'
                  )
fig.show()
```

```python
#population_df.diff()

fig=px.line(population_df[['India','China']],
        title='Population'  )

fig.update_layout(yaxis_title='Population_Count',
                  legend_title='Countries',
                  font_size=14,
                  legend_itemclick='toggle',
                  plot_bgcolor='orange',
                  legend_xref='paper',
                  font=dict(
                      family='Ariel',
                      size=14 ) )

fig.update_yaxes(rangemode='tozero');

fig.show()
```

```python
# TreeMap and Sunburst

gapminder_df=px.data.gapminder().query("year==2007")
gapminder_df["world"]="World"
gapminder_df
```

```python
# treemap
fig=px.treemap(gapminder_df,
               path=['world','continent','country'],
               values='pop',
               color='lifeExp',
               color_continuous_scale='RdBu' )

fig.show()

```

```python
# now replacing the treemap with sunburst
# sunburst
fig=px.sunburst(gapminder_df,
                path=['continent','country'],
                values='pop',
                color='lifeExp',
                color_continuous_scale='RdBu'
 )
fig.show()
```
